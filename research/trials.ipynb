```python
## Ensure annotation 
def get_predict(x:int, y:int) -> int:
    return x * y
                
```

```python
get_predict(x=2, y=3)
```

```python
# as if multiply int with string so it is multiplying 3 two times
get_predict(x=2, y="3")
```

```python
from ensure import ensure_annotations
```

```python
# so to avoid such small bugs that are not easily identifiable in large codes we do use ensure annotation 
@ensure_annotations
def get_predict(x:int, y:int) -> int:
    return x * y 
```

```python
get_predict(x=2, y=3)
```

```python
get_predict(x=2, y="3")
```
